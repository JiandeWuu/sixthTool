```python
import numpy as np
import pandas as pd
```

```python
enst2refseq = pd.read_csv("data/ucsc/enst2refseq.csv", sep="\t")
enst2refseq =enst2refseq.rename(columns={"hg38.kgXref.refseq": "refseq", "#hg38.knownGene.name": "ENST_version"})
enst2refseq['ENST'], _ = enst2refseq['ENST_version'].str.split('.').str
enst2refseq = enst2refseq.dropna()
```

```python
rbp_output = pd.read_csv("data/RBP-related/output.Ensembl.withAdjPval", sep="\t")
```

```python
refseq_0320 = pd.read_csv("data/society/cdhit80_0320_loc75_NC_SeqID.csv")
refseq_0320 = refseq_0320.rename(columns={"Seq_ID": "refseq_version"})[["refseq_version"]]
refseq_0320['refseq'], _ = refseq_0320['refseq_version'].str.split('.').str
```

```python
enst2refseq_0320 = pd.merge(enst2refseq, refseq_0320, on="refseq", how="inner")
print(len(refseq_0320.drop_duplicates()))
print(len(enst2refseq_0320[["refseq"]].drop_duplicates()))
print(enst2refseq_0320.head(5))
```

```python
df = rbp_output[rbp_output["AdjPval(-Log10)"] >= 2]
rbp2enst = []

for idx in df.index:
    rbp2enst.append([df["RBP|Cell|Rep"][idx].split("|")[0], df["LncRNA"][idx].split("|")[1], df["AdjPval(-Log10)"][idx]])
```

```python
rbp2enst_df = pd.DataFrame(rbp2enst, columns=["RBP", "ENST", "AdjPval(-Log10)"])
```

```python
rbp2refseq_df = pd.merge(rbp2enst_df, enst2refseq_0320, on="ENST", how="inner")
```

```python
rbp2refseq_df[["refseq_version"]].drop_duplicates()
```

```python
rbp2refseq_onehot_df = pd.DataFrame([],columns=rbp2refseq_df["RBP"].drop_duplicates(), index=rbp2refseq_df["refseq_version"].drop_duplicates())
# rbp2refseq_onehot_df.index = rbp2refseq_df["refseq_version"].drop_duplicates()
rbp2refseq_onehot_df.head(10)
```

```python
for idx in rbp2refseq_df[["RBP", "refseq_version"]].drop_duplicates().index:
    rbp2refseq_onehot_df[rbp2refseq_df["RBP"][idx]][rbp2refseq_df["refseq_version"][idx]] = 1
    # print(rbp2refseq_onehot_df[rbp2refseq_df["RBP"][idx]][rbp2refseq_df["refseq_version"][idx]])
    # print(rbp2refseq_df["RBP"][idx], rbp2refseq_df["refseq_version"][idx])
```

```python
rbp2refseq_onehot_df = rbp2refseq_onehot_df.fillna(0)
```

```python
rbp2refseq_onehot_df.head(10)
```

```python
rbp2refseq_onehot_df.to_csv("data/RBP-related/refseq2rbp_onehot.csv")
```
