```python
import numpy as np
import pandas as pd
```

```python
with open("data/Chiu/benchmark/benchmark_mRNA_nuc_v1.clstr") as f:
    arrays = f.read().split(">Cluster")
    cluster_array = []
    for cluster in arrays[1:]:
        cluster_dict = dict()
        cluster_dict["other"] = []
        for row in cluster.split("\n")[1:-1]:
            if "*" in row:
                cluster_array.append(row[row.index(">") + 1:-5])

with open("data/Chiu/benchmark/benchmark_mRNA_nuc_v1") as f:
    arrays = f.read().split(">")[1:]
    cluster_seq = []
    for i in range(len(arrays)):
    #     arrays[i] = arrays[i].split("\n")
        if arrays[i][:arrays[i].index("\n")] in cluster_array:
            cluster_seq.append(arrays[i].split("\n")[1])

nuc_df = pd.DataFrame(cluster_seq, columns=["seq"])
nuc_df["loc"] = "N"
nuc_df.head(10)
```

```python
with open("data/Chiu/benchmark/benchmark_mRNA_cyto_v1.clstr") as f:
    arrays = f.read().split(">Cluster")
    cluster_array = []
    for cluster in arrays[1:]:
        cluster_dict = dict()
        cluster_dict["other"] = []
        for row in cluster.split("\n")[1:-1]:
            if "*" in row:
                cluster_array.append(row[row.index(">") + 1:-5])

with open("data/Chiu/benchmark/benchmark_mRNA_cyto_v1") as f:
    arrays = f.read().split(">")[1:]
    cluster_seq = []
    for i in range(len(arrays)):
    #     arrays[i] = arrays[i].split("\n")
        if arrays[i][:arrays[i].index("\n")] in cluster_array:
            cluster_seq.append(arrays[i].split("\n")[1])

cyto_df = pd.DataFrame(cluster_seq, columns=["seq"])
cyto_df["loc"] = "C"
cyto_df.head(10)
```

```python
import re
with open("data/Chiu/benchmark/benchmark_mRNA_nuc_v1.clstr") as f:
    arrays = f.read().split(">Cluster")
    cluster_array = []
    for cluster in arrays[1:]:
        cluster_dict = dict()
        cluster_dict["other"] = []
        for row in cluster.split("\n")[1:-1]:
            if "*" in row:
                cluster_dict["main"] = row[row.index(">")+1:row.index("|")]
            cluster_dict["other"].append(row[row.index(">")+1:row.index("|")])
        cluster_array.append(cluster_dict)
```

```python
cluster_array
```

```python
rep2seq = []
for cluster in cluster_array:
    represent_refseq = cluster["main"]
    for refseq in cluster["other"]:
        rep2seq.append([represent_refseq, refseq])
df = pd.DataFrame(rep2seq, columns=["Represent_Refseq", "Sub_Refseq"])
df.head(10)
```

```python
df.to_csv("data/society/cdhit0320_RepresentRefseq2SubRefseq.csv", index=False)
```
